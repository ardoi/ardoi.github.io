```python
points = 500
t=arange(0, points)
noise = random.randn(points)*300
fit = 4000*sin(t/40.)**2*exp(-t/200.)
raw = fit + noise
```

```python
plot(t, fit, label='Fit')
plot(t, raw,'o', label='Raw data')
legend()
ylabel('Raw fluorescence')
xlabel('Time')
savefig('img0.png',dpi=150, bbox_inches="tight")
```

```python
plot(t, raw, 'o', label='Raw data', color='#0EA16A', alpha=0.75)
plot(t, fit, '#EF5915', lw=3, label='Fit')
legend()
ylabel('Raw fluorescence')
xlabel('Time')
savefig('img1.png',dpi=150, bbox_inches="tight")
```

```python
plot(t, raw, 'o', label='Raw data', color='#0EA16A', mec='None', alpha=0.75,  ms=5)
plot(t, fit, '#EF5915', lw=3, label='Fit')
legend()
ylabel('Raw fluorescence')
xlabel('Time')
savefig('img2.png',dpi=150, bbox_inches="tight")
```

```python
def myfig():
    fig = plt.figure()
    ax = fig.gca()
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')
    return fig, ax
```

```python
fig,ax = myfig()
ax.plot(t, raw, 'o', label='Raw data', color='#0EA16A', mec='None', alpha=0.75,  ms=5)
ax.plot(t,fit,'#EF5915',lw=3,label='Fit')
ax.legend()
ylabel('Raw fluorescence')
xlabel('Time')
savefig('img3.png',dpi=150, bbox_inches="tight")
```
